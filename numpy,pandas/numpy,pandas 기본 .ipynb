```python
#cabin ,Embarked
df['Cabin'].str[0].value_counts()
df['Cabin']=df['Cabin'].fillna('C00') # 널 처리
```

```python
df['Embarked'].value_counts()
df['Embarked'].fillna('S',inplace=True)# null 처리
```

```python
df['Age'].value_counts() #나이를 그룹핑해서 데이터를 들여다 봅시다

#10~20대, 30~40대, 고령자
df['Age']<=20
df['Age']<=40
df['Age']>40
```

```python
def age_group(age):
    if age <= 20:
        group = 'Student'
    elif age <= 40:
        group = 'Adult'
    else:
        group = 'Senior'
    return group

df['Age_Group'] = df['Age'].apply(age_group)
```

```python
#나이의 그룹에 따른 생존확률
df.groupby('Age_Group')['Survived'].mean()
```

```python
#60대 이상의 고령자는 생존 확률이 어떻게 되었을까?

df_age_high=df[df['Age']>=60][['Survived','Age','Sex','Fare','Embarked']]
df_age_high.describe()
```

```python
# 클래스 별 생존확률
ax = df.groupby('Pclass')['Survived'].mean().sort_values(ascending=True).plot()
ax.set_ylabel('Survived')

```

```python
ax = df.groupby('Pclass')['Survived'].mean().sort_values(ascending=True).plot(kind='bar')
ax.set_ylabel('Survived')
```
