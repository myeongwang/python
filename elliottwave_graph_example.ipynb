```python
import matplotlib.pyplot as plt
import numpy as np

# 예시 데이터 생성
x = np.linspace(0, 10, 1000)
impulse_wave = [0, 2, 1, 3, 2.5, 4]
corrective_wave = [3.5, 3.7, 3.2]

# x 좌표를 위한 리스트
impulse_x = [0, 2, 4, 6, 8, 10]
corrective_x = [10, 12, 14]

# 그래프 그리기
plt.figure(figsize=(10,6))
plt.plot(impulse_x, impulse_wave, label="Impulse Wave", color='blue')
plt.plot(corrective_x, corrective_wave, label="Corrective Wave", color='red', linestyle="--")
plt.axhline(y=impulse_wave[-1], color='green', linestyle='--')  # 5파의 고점을 나타내는 수평선

# 라벨 및 제목
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Elliott Wave Principle")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 파동 데이터
impulse_x = [1, 2, 3, 4, 5, 6]
impulse_wave = [1, 3, 2, 4, 3, 5]

corrective_x = [6, 7, 8, 9]
corrective_wave = [5, 4, 4.5, 3.5]

# 그래프 그리기
plt.figure(figsize=(10,6))

# Impulse Wave
plt.plot(impulse_x, impulse_wave, label="Impulse Wave", color='blue', marker='o')
for i, txt in enumerate(['1', '2', '3', '4', '5']):
    plt.annotate(txt, (impulse_x[i], impulse_wave[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Corrective Wave
plt.plot(corrective_x, corrective_wave, label="Corrective Wave", color='red', linestyle='--', marker='o')
for i, txt in enumerate(['A', 'B', 'C']):
    plt.annotate(txt, (corrective_x[i], corrective_wave[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.axhline(y=impulse_wave[-1], color='green', linestyle='--', label="Wave 5 Peak")  # 5파의 고점을 나타내는 수평선

# 라벨 및 제목
plt.xlabel("Wave Number")
plt.ylabel("Value")
plt.title("Elliott Wave Principle")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def impulse_subwave(start):
    return [start, start + 0.5, start - 0.2, start + 0.3, start - 0.1, start + 1]

def corrective_subwave(start):
    return [start, start - 0.3, start + 0.2, start - 0.5]

# 파동 데이터
waves = np.array([
    *impulse_subwave(1), 
    *corrective_subwave(4.5),
    *impulse_subwave(3.7),
    *corrective_subwave(8.8),
    *impulse_subwave(7.5),
    *impulse_subwave(14),
    *corrective_subwave(13),
    *corrective_subwave(9.2),
])

x_vals = list(range(len(waves)))

# 그래프 그리기
plt.figure(figsize=(15, 8))

plt.plot(x_vals, waves, label="Wave with Subwaves", color='blue', marker='o')

# 라벨 및 제목
plt.xlabel("Subwave Number")
plt.ylabel("Value")
plt.title("Elliott Wave with Subwaves")
plt.axhline(y=waves[10], color='red', linestyle='--', label="Wave 5 Peak")  # 5파의 고점을 나타내는 수평선

# 주요 파동 위치 라벨 추가
major_wave_points = [5, 8, 13, 16, 21, 26, 29, 32]
major_wave_labels = ['1', '2', '3', '4', '5', 'A', 'B', 'C']

for i, point in enumerate(major_wave_points):
    plt.annotate(major_wave_labels[i], (x_vals[point], waves[point]), textcoords="offset points", xytext=(0,10), ha='center')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 파동 데이터
def impulse_subwave(start):
    return [start, start + 1, start + 0.8, start + 2, start + 1.8, start + 3]

def corrective_subwave(start):
    return [start, start - 1, start - 0.5, start - 1.5]

impulse_wave = impulse_subwave(1) + corrective_subwave(impulse_subwave(1)[-1]) + \
               impulse_subwave(corrective_subwave(impulse_subwave(1)[-1])[-1]) + \
               corrective_subwave(impulse_subwave(corrective_subwave(impulse_subwave(1)[-1])[-1])[-1]) + \
               impulse_subwave(corrective_subwave(impulse_subwave(corrective_subwave(impulse_subwave(1)[-1])[-1])[-1])[-1]) + \
               [corrective_subwave(impulse_subwave(corrective_subwave(impulse_subwave(corrective_subwave(impulse_subwave(1)[-1])[-1])[-1])[-1])[-1])[-1] - 3]

corrective_wave = [impulse_subwave(corrective_subwave(impulse_subwave(corrective_subwave(impulse_subwave(1)[-1])[-1])[-1])[-1])[-1] - 3,
                  impulse_subwave(corrective_subwave(impulse_subwave(corrective_subwave(impulse_subwave(1)[-1])[-1])[-1])[-1])[-1] - 1,
                  impulse_subwave(corrective_subwave(impulse_subwave(corrective_subwave(impulse_subwave(1)[-1])[-1])[-1])[-1])[-1]]

x_vals = list(range(1, len(impulse_wave) + len(corrective_wave) + 1))

# 그래프 그리기
plt.figure(figsize=(10,6))

# Impulse Wave with subwaves
plt.plot(x_vals, impulse_wave + corrective_wave, label="Impulse & Corrective Wave", color='blue', marker='o')
major_wave_labels = ['1', '2', '3', '4', '5', 'A', 'B', 'C']
major_wave_points = [5, 8, 13, 16, 21, 26, 27, 28]

for i, point in enumerate(major_wave_points):
    plt.annotate(major_wave_labels[i], (x_vals[point-1], impulse_wave[point-1] if point-1 < len(impulse_wave) else corrective_wave[point-1-len(impulse_wave)]), textcoords="offset points", xytext=(0,10), ha='center')

plt.axhline(y=impulse_wave[21], color='green', linestyle='--', label="Wave 5 Peak")  # 5파의 고점을 나타내는 수평선

# 라벨 및 제목
plt.xlabel("Subwave Number")
plt.ylabel("Value")
plt.title("Elliott Wave with Subwaves")
plt.legend()
plt.grid(True)
plt.show()
```
