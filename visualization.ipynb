```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# plot(): 선(라인) 그래프 
plt.plot([1,2,3,4,5,4,3,2,1])
```

```python
x= np.arange(0,10,1)
y=np.sin(x)
y2 = np.cos(x)

plt.figure(facecolor='red',figsize=(12,8),dpi=600) #도화지 옵션 #dpi 높이면 해상도 올라감 확대해도 안깨짐 

plt.plot(x,y,label='sin',marker='o',markersize=6,markeredgecolor ='red')
plt.plot(x,y2,label='cos',marker='x',markersize=10,linestyle='-.')
plt.legend(loc=(1.05,0)) # 범례 추가 

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sin/Cos Graph')
plt.grid() #격자 추가 
```

```python
#그래프를 여러 개를 따로 나타내는 방법 => subplots()

y3=x*2
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=2,figsize=(8,6),dpi=600)
ax1.plot(x,y3)
ax2.bar(x,y3)
ax3.plot(x,y3)
ax4.bar(x,y3)
```

```python
def age_group(age):
    if age <=10: group='baby'
    elif age<=20: group='student'
    elif age<=30: group='young adult'
    elif age<=40: group='adult'
    else: group ='senior'
    
    return group 
```

```python
df['Age_Group']=df['Age'].apply(age_group)
orders=['baby','student','young adult','adult','senior']
sns.barplot(data=df,x='Age_Group',y='Survived',order=orders,hue='Sex')
```

```python
#여성과 남성 각각에 대한 나이 그룹을 보고 싶어요.
plt.figure(figsize=(20,12))
sns.barplot(data=df,x='Sex',y='Survived',hue='Age_Group')
```
